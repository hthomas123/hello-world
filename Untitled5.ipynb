```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import accuracy_score

# Assuming 'shuffled_iris_df' DataFrame is already defined and contains the shuffled Iris dataset

# Create a new DataFrame with the first 130 entries of 'shuffled_iris_df'
subset_iris = shuffled_iris_df.iloc[:130].copy()

# Extract sepal length and petal width as features
features = subset_iris[['sepal_length', 'petal_width']]

# Perform k-Means Clustering with three clusters
kmeans = KMeans(n_clusters=3)
kmeans.fit(features)

# Add cluster labels to the DataFrame
subset_iris['cluster'] = kmeans.labels_

# Plot clusters including centroids
plt.figure(figsize=(8, 6))
sns.scatterplot(data=subset_iris, x='sepal_length', y='petal_width', hue='cluster', palette='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], marker='x', s=200, c='red', label='Centroids')
plt.title('Clusters with Centroids')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.legend(title='Cluster')
plt.show()

# Use the remaining 20 entries of the shuffled dataset to estimate assignment error
test_iris = shuffled_iris_df.iloc[130:].copy()
test_features = test_iris[['sepal_length', 'petal_width']]
test_iris['predicted_cluster'] = kmeans.predict(test_features)

# Calculate assignment error
true_labels = test_iris['species']  # Assuming 'species' column exists in the dataset
predicted_labels = test_iris['predicted_cluster']
assignment_error = 1 - accuracy_score(true_labels, predicted_labels)

print(f"Assignment Error on Unseen Data: {assignment_error * 100:.2f}%")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
import numpy as np

# Load the Iris dataset into a DataFrame using Seaborn
iris_df = sns.load_dataset('iris')

# Label setosa as -1 and the other two species as +1
iris_df['label'] = np.where(iris_df['species'] == 'setosa', -1, 1)

# Print the DataFrame to verify labeling
print(iris_df.head())

# Select features and labels
X = iris_df[['sepal_length', 'petal_width']]
y = iris_df['label']

# Fit Perceptron model
perceptron = Perceptron()
perceptron.fit(X, y)

# Plot the decision boundary
w = perceptron.coef_[0]
b = perceptron.intercept_[0]

x_min, x_max = X.iloc[:, 0].min() - 1, X.iloc[:, 0].max() + 1
y_min, y_max = X.iloc[:, 1].min() - 1, X.iloc[:, 1].max() + 1

xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.02),
                     np.arange(y_min, y_max, 0.02))

Z = perceptron.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

plt.contourf(xx, yy, Z, cmap=plt.cm.coolwarm, alpha=0.8)
plt.scatter(X.iloc[:, 0], X.iloc[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel('Sepal Length')
plt.ylabel('Petal Width')
plt.title('Perceptron Separator')
plt.show()
```
